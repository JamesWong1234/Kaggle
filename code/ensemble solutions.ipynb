```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

base_path = '../input/'

# test files
test_xgb = pd.read_csv(base_path + 'Fork_of_ForK_LightGBM_with_Simple_Features.csv')
test_lgb = pd.read_csv(base_path + 'submission_bayesian_optimization.csv')
test_dnn = pd.read_csv(base_path + 'submission_bayesian_optimization_791.csv')
test_up = pd.read_csv(base_path + 'submission_bayesian_optimization_0792.csv')
test_cat = pd.read_csv(base_path + 'submission_random_search_790.csv')
test_kin = pd.read_csv(base_path + 'submission_with selected_features.csv')
test_gp = pd.read_csv(base_path + 'tidy_xgb_0.78889.csv')
test=pd.read_csv(base_path + 'application_test.csv')


```

```python
test_xgb.columns
```

```python
test = pd.concat([test, 
                   test_xgb[['TARGET']].rename(columns = {'TARGET' : 'LightGBM'}),
                   test_lgb[['TARGET']].rename(columns = {'TARGET' : 'bayesian'}),
                   test_dnn[['TARGET']].rename(columns = {'TARGET' : 'bayesian791'}),
                   test_up[['TARGET']].rename(columns = {'TARGET' : 'bayesian792'}),
                   test_cat[['TARGET']].rename(columns = {'TARGET' : 'random_search_790'}),
                   test_kin[['TARGET']].rename(columns = {'TARGET' : 'selected_features'}),
                   test_gp[['TARGET']].rename(columns = {'TARGET' : 'tidy_xgb'})                   
                  ], axis = 1)


print(test.head(10))
```

```python

train_cols = ['LightGBM', 'bayesian', 'bayesian791', 'bayesian792', 'random_search_790', 'selected_features', 'tidy_xgb']
for t in train_cols:
    test[t + '_rank'] = test[t].rank()
```

```python
test['TARGET'] = (test['LightGBM'] + test['bayesian'] + test['bayesian791'] + test['bayesian792'] + \
                 test['random_search_790'] + test['selected_features'] + test['tidy_xgb']) / (7 * test.shape[0])

```
